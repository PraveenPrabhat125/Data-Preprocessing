# Dealing With Categorical Values

## Importing Libraries

```python
import pandas as pd
import numpy as np
```

## Importing Dataset

```python
dataset = pd.read_csv("Salary_Dataset.csv")
```

```python
dataset.head(40)
```

# Lets perform Encoding

## Applying One-Hot Encoding

<h3> First Do it with pandas </h3>

```python
# Get dummies with pandas in any variable
country_dummy = pd.get_dummies(dataset['country'])
country_dummy
```

```python
dataset.columns
```

```python
lis = ['Australia','Canada', 'Dubai', 'USA']
dataset.drop(lis, axis = 1, inplace = True)
# concatenate the dataframes into original dataframes.
# axis = 1 as to add columns wise
```

```python
df = pd.concat([dataset, country_dummy], axis = 1)
df.head(10)
```

```python
# Drop the country column and rearrange the index of column.

df.drop(['country'], axis = 1, inplace = True)

df
```

```python
df.columns
# df = df[['']]
```

```python
df1 = df[['Australia', 'Canada',
       'Dubai', 'USA', 'Salary', 'YearsExperience', 'Purchased']]
df1
```

# <h3> Let's do it with Scikit-Learn </h3>
Also Restart your kernel if you are using same notebook


```python
# Lets first perform label encoding
# Labelencoding will be performed on Purchased column as it has only two unique value in it.

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() # this is instance
```

```python
# Lets perform one-hot encoding on country column.
df1.head(5)
```

```python
# label encoding
df1['Purchased'] = le.fit_transform(df1['Purchased'])
```

# One hot encoding

```python
dataset = pd.read_csv('/content/Salary_Dataset.csv')
dataset.head(5)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers= [('encoder', OneHotEncoder(), [0]) ] , remainder = 'passthrough')

dataset = ct.fit_transform(dataset)
```

```python
#Lets link the headers name.
type(dataset)
dataset
# hot encoding return nd array
df = pd.DataFrame(dataset)
df[:10]
```

```python

df.columns = [['Australia',	'Canada', 	'Dubai', 	'USA'	, 'Salary', 	'YearsExperience', 	'Purchased']]
df
```

```python
df.columns
```

```python
# label encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Purchased'] = le.fit_transform(df['Purchased'])
```
