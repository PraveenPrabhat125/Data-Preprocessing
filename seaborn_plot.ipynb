```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('diamonds')
df.head(10)
```

```python
df.columns
```

```python
sns.countplot(x = 'cut', data = df, palette = 'bright')
```

```python
df.columns
```

```python
price_cut = df.groupby(['cut'], as_index = False).agg({
    'price': 'sum'
})
price_cut
```

```python
ax = sns.barplot(data = price_cut, x = 'cut', y = 'price', palette = 'bright')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
price_cut.head()
```

```python
plt.pie(price_cut['price'], labels = price_cut['cut'], data = price_cut, autopct= "%1.2f%%" )
```

# note:
1. price, carat is right skewed hvaing outliers at right side
2. depth and table are normal distributed but having some outliers on both side

```python
qh25, qh50, qh75 = np.percentile(df['price'], [25,50, 75])
print(qh25, qh50, qh75)


iqrh = qh75 - qh25

maxh =  qh75 + 0.9*iqrh

df_price = df[df['price']< maxh]


sns.boxplot(data = df_price, x = 'price', color = 'r')
```

```python
sns.histplot(data = df_price, x = 'price')
```

```python
# Depth

qh25, qh50, qh75  = np.percentile(df['depth'], [25, 50, 75])

iqrh = qh75 - qh25

minh = qh25 - 1.4*iqrh
maxh = qh75 + 1.4*iqrh

df_depth = df[(df['depth']> minh) & (df['depth']<maxh)]

sns.boxplot(data = df_depth, x = 'depth', color = 'b')

```

```python
sns.histplot(data = df_depth, x = 'depth', color = 'b')
```
